```python
import os
import re
from pathlib import Path
import numpy as np
from patchify import patchify
from PIL import Image
import cv2
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#creating empty folders
def create_folders():
    Folders = ['train', 'val', 'test']
    SubFolders = ['images', 'mask']

    for folder in Folders:
        if not os.path.exists(folder):
            for subfolder in SubFolders:
                os.makedirs(folder+'/'+subfolder) if not os.path.exists(folder+'/'+subfolder) else print(f'{folder} or {subfolder} already exist')
```

```python
create_folders()
```

```python
def create_patches(src, dst):
    path_split = os.path.split(src)
    tile_num = re.findall(r'\d+', path_split[0])[0]

    img = cv2.imread(src)
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    
    image = np.asarray(img)

    if image.shape[2] > 2:
        patches = patchify(image, (320,320,3), step=300)
        file_name_wo_ext = Path(src).stem

        for i in range(patches.shape[0]):
            for j in range(patches.shape[1]):
                patch = patches[i,j,0]
                patch = Image.fromarray(patch)
                num =  i * patches.shape[1] + j
                patch.save(f"{dst}/{file_name_wo_ext}_tile_{tile_num}_patch_{num}.png")
```

```python
for path_name, _, file_name in os.walk('data'):
    for f in file_name:
        if f != 'classes.json':
            path_split = os.path.split(path_name)
            tile_num = re.findall(r'\d+', path_split[0])[0]

            img_type = path_split[1]

            if tile_num == '1':
                target_folder_img = target_folder_mask = 'test'
            
            elif tile_num == '3':
                target_folder_img = target_folder_mask = 'val'

            elif tile_num in ['4','5','6','7','8']:
                target_folder_img = target_folder_mask = 'train'
            
            src = os.path.join(path_name, f)
            file_name_wo_ext = Path(src).stem

            img_file = f'{path_split[0]}/images/{file_name_wo_ext}.jpg'
            mask_file = f'{path_split[0]}/mask/{file_name_wo_ext}.png'

            if os.path.exists(img_file) and os.path.exists(mask_file):
                if img_type == 'images':
                    dst = os.path.join(target_folder_img,img_type)
                    create_patches(src=src,dst=dst)
                
                if img_type == 'mask':
                    dst = os.path.join(target_folder_mask,img_type)
                    create_patches(src=src,dst=dst)

        

            


```
